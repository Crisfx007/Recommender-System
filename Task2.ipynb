```python
### We will be using two-tower approach 

# User tower features (14)
user_cat_cols = [
    "Customer_ID", "City", "State", "Country", "Gender",
    "Income", "Customer_Segment", "Feedback",
    "Shipping_Method", "Payment_Method", "Order_Status"
]
user_num_cols = ["Age", "Total_Purchases", "Amount"]

# Item tower features (4)
item_cat_cols = ["products", "Product_Category", "Product_Brand", "Product_Type"]

# Target and date
target_col = "Ratings"
date_col = "Date"
```

```python
df_clean = df_clean[df_clean["Order_Status"] == "Delivered"].copy()
print("Rows after keeping only Delivered orders:", len(df_clean))
```

```python
import numpy as np
all_cat_cols = user_cat_cols + item_cat_cols
vocab_sizes = {}
mappings = {}

for col in all_cat_cols:
    df_clean[col] = df_clean[col].astype("category")
    mapping = {val: idx for idx, val in enumerate(df_clean[col].cat.categories)}
    df_clean[col + "_idx"] = df_clean[col].cat.codes.astype(np.int32)
    vocab_sizes[col] = len(df_clean[col].cat.categories)
    mappings[col] = mapping

print("Vocab sizes:", vocab_sizes)
```

```python
from sklearn.preprocessing import StandardScaler

scalers = {}

# Age → standard scale
scalers["Age"] = StandardScaler().fit(df_clean[["Age"]])
df_clean["Age_s"] = scalers["Age"].transform(df_clean[["Age"]])

# Total_Purchases → min-max 0–1
min_tp, max_tp = df_clean["Total_Purchases"].min(), df_clean["Total_Purchases"].max()
df_clean["Total_Purchases_s"] = (df_clean["Total_Purchases"] - min_tp) / (max_tp - min_tp + 1e-9)

# Amount → log1p + standard scale
df_clean["Amount_log"] = np.log1p(df_clean["Amount"])
scalers["Amount_log"] = StandardScaler().fit(df_clean[["Amount_log"]])
df_clean["Amount_log_s"] = scalers["Amount_log"].transform(df_clean[["Amount_log"]])
```

```python
df_clean[date_col] = pd.to_datetime(df_clean[date_col])
df_clean = df_clean.sort_values(date_col)

train_cut = df_clean[date_col].quantile(0.70)
val_cut = df_clean[date_col].quantile(0.85)

train_df = df_clean[df_clean[date_col] <= train_cut].copy()
val_df = df_clean[(df_clean[date_col] > train_cut) & (df_clean[date_col] <= val_cut)].copy()
test_df = df_clean[df_clean[date_col] > val_cut].copy()

print("Train / Val / Test sizes:", len(train_df), len(val_df), len(test_df))
```

```python
from collections import defaultdict

user_pos = defaultdict(set)
for u, item in zip(train_df["Customer_ID_idx"], train_df["products_idx"]):
    user_pos[int(u)].add(int(item))

num_users = len(train_df["Customer_ID_idx"].unique())
num_items = len(train_df["products_idx"].unique())
print("Num users:", num_users, "Num items:", num_items)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import random
from tqdm import tqdm
import numpy as np

class TwoTowerDataset(Dataset):
    def __init__(self, df_clean, user_cat_cols, user_num_cols, item_cat_cols, target_col,
                 user_pos=None, num_negatives=8, mode="train"):
        self.df_clean = df_clean.reset_index(drop=True)
        self.user_cat_cols = [c + "_idx" for c in user_cat_cols]
        
        self.user_num_cols = []
        for c in user_num_cols:
            if c == "Age":
                self.user_num_cols.append("Age_s")
            elif c == "Total_Purchases":
                self.user_num_cols.append("Total_Purchases_s")  
            elif c == "Amount":
                self.user_num_cols.append("Amount_log_s")
            else:
                self.user_num_cols.append(c)
                
        self.item_cat_cols = [c + "_idx" for c in item_cat_cols]
        self.target_col = target_col
        self.user_pos = user_pos
        self.num_negatives = num_negatives
        self.mode = mode

        self.user_feat_dict = {}
        for user, group in df_clean.groupby("Customer_ID_idx"):
            self.user_feat_dict[user] = {
                'cat': torch.tensor(group[self.user_cat_cols].iloc[0].values, dtype=torch.long),
                'num': torch.tensor(group[self.user_num_cols].iloc[0].values, dtype=torch.float)
            }

        self.item_feat_dict = {}
        for item, group in df_clean.groupby("products_idx"):
            self.item_feat_dict[item] = {
                'cat': torch.tensor(group[self.item_cat_cols].iloc[0].values, dtype=torch.long)
            }

        self.samples = []
        all_items = set(self.item_feat_dict.keys())
        
        if mode == "train":
            for _, row in df_clean.iterrows():
                user = row["Customer_ID_idx"]
                item = row["products_idx"]
                rating = row[target_col]

                normalized_rating = (rating - 1.0) / 4.0  
                self.samples.append((user, item, rating, normalized_rating))


                if user_pos and user in user_pos:
                    neg_candidates = list(all_items - user_pos[user])
                else:
                    neg_candidates = list(all_items - {item})
                
                if neg_candidates:
                    neg_items = random.sample(neg_candidates, min(self.num_negatives, len(neg_candidates)))
                    for neg in neg_items:
                        
                        neg_rating = random.uniform(0.1, 0.3) 
                        self.samples.append((user, neg, neg_rating * 4 + 1, neg_rating))
        else:
            
            for _, row in df_clean.iterrows():
                user = row["Customer_ID_idx"]
                item = row["products_idx"]
                rating = row[target_col]
                normalized_rating = (rating - 1.0) / 4.0
                self.samples.append((user, item, rating, normalized_rating))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        user, item, original_rating, normalized_rating = self.samples[idx]
        
        user_cat = self.user_feat_dict[user]['cat']
        user_num = self.user_feat_dict[user]['num']
        item_cat = self.item_feat_dict[item]['cat']

        return user_cat, user_num, item_cat, torch.tensor(normalized_rating, dtype=torch.float)

class TwoTowerHybrid(nn.Module):
    def __init__(self, user_cat_sizes, user_num_dims, item_cat_sizes, emb_dim=64, hidden_dims=[128, 64]):
        super().__init__()
        

        self.user_embs = nn.ModuleDict({
            col: nn.Embedding(size, emb_dim) for col, size in user_cat_sizes.items()
        })
        

        user_input_dim = len(user_cat_sizes) * emb_dim + user_num_dims
        self.user_fc = nn.Sequential(
            nn.Linear(user_input_dim, hidden_dims[0]),
            nn.ReLU(),
            nn.Dropout(0.1), 
            nn.Linear(hidden_dims[0], hidden_dims[1]),
            nn.ReLU(),
            nn.Dropout(0.1)
        )

        
        self.item_embs = nn.ModuleDict({
            col: nn.Embedding(size, emb_dim) for col, size in item_cat_sizes.items()
        })
        
        item_input_dim = len(item_cat_sizes) * emb_dim
        self.item_fc = nn.Sequential(
            nn.Linear(item_input_dim, hidden_dims[1]),
            nn.ReLU(),
            nn.Dropout(0.1)
        )
        
        
        self.global_bias = nn.Parameter(torch.zeros(1))
        
        
        self.apply(self._init_weights)

    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            nn.init.xavier_uniform_(module.weight)
            if module.bias is not None:
                nn.init.constant_(module.bias, 0)
        elif isinstance(module, nn.Embedding):
            nn.init.xavier_uniform_(module.weight)  

    def forward(self, user_cat, user_num, item_cat):
        # User tower
        user_emb_list = [self.user_embs[col](user_cat[:, i]) for i, col in enumerate(self.user_embs.keys())]
        user_emb = torch.cat(user_emb_list + [user_num], dim=1)
        user_vec = self.user_fc(user_emb)

        # Item tower
        item_emb_list = [self.item_embs[col](item_cat[:, i]) for i, col in enumerate(self.item_embs.keys())]
        item_emb = torch.cat(item_emb_list, dim=1)
        item_vec = self.item_fc(item_emb)

        
        user_vec_norm = F.normalize(user_vec, p=2, dim=1)
        item_vec_norm = F.normalize(item_vec, p=2, dim=1)
        
        # Cosine similarity + bias + scaling
        cosine_sim = (user_vec_norm * item_vec_norm).sum(dim=1)
        score = cosine_sim + self.global_bias
        
        # Scale to [0,1] range for rating prediction
        score = torch.sigmoid(score)
        return score


def train_two_tower_model(train_dataset, val_dataset, user_cat_sizes, item_cat_sizes, user_num_cols,
                         batch_size=512, epochs=15, lr=3e-3, device=None):
    
    if device is None:
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=0)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=0)
    
    model = TwoTowerHybrid(
        user_cat_sizes=user_cat_sizes,
        user_num_dims=len(user_num_cols),
        item_cat_sizes=item_cat_sizes,
        emb_dim=64,  
        hidden_dims=[128, 64]  
    ).to(device)
    
    
    criterion = nn.MSELoss()
    

    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-6)
    

    scheduler = optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='min', factor=0.7, patience=3
    )
    
    print(f"Training on device: {device}")
    print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")
    print(f"Training samples: {len(train_dataset)}, Validation samples: {len(val_dataset)}")
    
    best_val_loss = float('inf')
    patience = 6
    patience_counter = 0
    
    for epoch in range(epochs):
        # Training phase
        model.train()
        total_loss = 0
        num_batches = 0
        
        progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{epochs}')
        for user_cat, user_num, item_cat, rating in progress_bar:
            user_cat = user_cat.to(device)
            user_num = user_num.to(device) 
            item_cat = item_cat.to(device)
            rating = rating.to(device)
            
            optimizer.zero_grad()
            outputs = model(user_cat, user_num, item_cat)
            loss = criterion(outputs, rating)
            loss.backward()
            

            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            
            optimizer.step()
            
            total_loss += loss.item()
            num_batches += 1
            
            progress_bar.set_postfix({'loss': f'{loss.item():.4f}'})
        
        avg_train_loss = total_loss / num_batches
        
        # Validation phase
        model.eval()
        val_loss = 0
        num_val_batches = 0
        mae_total = 0  # Mean Absolute Error for interpretability
        
        with torch.no_grad():
            for user_cat, user_num, item_cat, rating in val_loader:
                user_cat = user_cat.to(device)
                user_num = user_num.to(device)
                item_cat = item_cat.to(device)
                rating = rating.to(device)
                
                outputs = model(user_cat, user_num, item_cat)
                loss = criterion(outputs, rating)
                val_loss += loss.item()
                
                pred_ratings = outputs * 4 + 1  # Convert 0-1 back to 1-5
                actual_ratings = rating * 4 + 1
                mae_total += torch.abs(pred_ratings - actual_ratings).sum().item()
                
                num_val_batches += 1
        
        avg_val_loss = val_loss / num_val_batches
        avg_mae = mae_total / len(val_dataset)
        
        # Learning rate scheduling
        scheduler.step(avg_val_loss)
        
        print(f"Epoch {epoch+1}/{epochs}")
        print(f"  Train Loss: {avg_train_loss:.4f}")
        print(f"  Val Loss: {avg_val_loss:.4f}, MAE: {avg_mae:.4f}")
        print(f"  Current LR: {optimizer.param_groups[0]['lr']:.6f}")
        print("-" * 50)
        
        # Early stopping
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            patience_counter = 0
            torch.save(model.state_dict(), 'best_two_tower_model.pth')
            print(f"  New best model saved! Val Loss: {best_val_loss:.4f}")
        else:
            patience_counter += 1
            
        if patience_counter >= patience:
            print(f"Early stopping triggered after {epoch+1} epochs")
            break
    
    # Load best model
    model.load_state_dict(torch.load('best_two_tower_model.pth'))
    print(f"Training completed. Best validation loss: {best_val_loss:.4f}")
    return model
```

```python
train_dataset = TwoTowerDataset(
    train_df, user_cat_cols, user_num_cols, item_cat_cols, 
    target_col, user_pos=user_pos, num_negatives=4, mode='train'
)

val_dataset = TwoTowerDataset(
    val_df, user_cat_cols, user_num_cols, item_cat_cols, 
    target_col, user_pos=user_pos, num_negatives=0, mode='val'
)

model = train_two_tower_model(
    train_dataset=train_dataset,
    val_dataset=val_dataset,
    user_cat_sizes=user_cat_sizes,
    item_cat_sizes=item_cat_sizes,
    user_num_cols=user_num_cols,
    batch_size=512,
    epochs=5,
    lr=3e-3
)
```

```python
def create_recommendation_assets(df_clean, save_path="recommendation_assets/"):

    os.makedirs(save_path, exist_ok=True)
    
    print("Creating recommendation assets...")
    
    # 1. Item catalog
    item_catalog = []
    seen_items = set()
    
    for _, row in df_clean.iterrows():
        item_key = row['products']
        if item_key not in seen_items:
            item_catalog.append({
                'products': row['products'],
                'Product_Category': row['Product_Category'],
                'Product_Brand': row['Product_Brand'],
                'Product_Type': row['Product_Type']
            })
            seen_items.add(item_key)
    
    # 2. User profiles
    user_profiles = {}
    for customer_id, group in df_clean.groupby('Customer_ID'):
        user_data = group.iloc[0]
        user_profiles[customer_id] = {
            'Customer_ID': user_data['Customer_ID'],
            'City': user_data['City'],
            'State': user_data['State'],
            'Country': user_data['Country'],
            'Gender': user_data['Gender'],
            'Income': user_data['Income'],
            'Customer_Segment': user_data['Customer_Segment'],
            'Feedback': user_data['Feedback'],
            'Shipping_Method': user_data['Shipping_Method'],
            'Payment_Method': user_data['Payment_Method'],
            'Order_Status': user_data['Order_Status'],
            'Age': user_data['Age'],
            'Total_Purchases': user_data['Total_Purchases'],
            'Amount': user_data['Amount']
        }
    
    # 3. User purchase histories
    user_histories = {}
    for customer_id, group in df_clean.groupby('Customer_ID'):
        user_histories[customer_id] = list(group['products'].unique())
    
    # 4. Rating statistics for scaling
    rating_stats = {
        'min_rating': df_clean['Ratings'].min(),
        'max_rating': df_clean['Ratings'].max(),
        'mean_rating': df_clean['Ratings'].mean()
    }
    
    # Save all assets
    assets = {
        'item_catalog': item_catalog,
        'user_profiles': user_profiles,
        'user_histories': user_histories,
        'rating_stats': rating_stats
    }
    
    with open(f"{save_path}/recommendation_assets.pkl", "wb") as f:
        pickle.dump(assets, f)
    
    print(f"Assets saved to {save_path}")
    print(f"- {len(item_catalog)} unique items")
    print(f"- {len(user_profiles)} user profiles")
    print(f"- {len(user_histories)} user purchase histories")
    
    return assets
```

```python

# This extracts everything needed from df_clean and saves it
import os
create_recommendation_assets(df_clean, "recommendation_assets/")
```
